```python
"""
articleNo : 매물 번호(primary key)
floorInfo : 해당층/총층
supplySpace : 공급면적
exclusiveSpace : 전용면적
roomCount : 방수
bathroomCount : 욕실수
monthlyManagementCost : 월 관리비
articleName : 아파트 이름과 동
direction : 남향
articleFeatureDes : 매물 특징
moveInPossibleYmd : 입주 가능일
parkingCount : 총 주차대수
parkingPerHouseholdCount : 1인당 주차대수
realtorName : 부동산 이름
representativeTelNo : 대표번호
realestateTypeName : 매물 유형
detailDescription : 매물 설명
tradeTypeName : 거래방식
warrantPrice : 보증금
dealPrice : 매매가
rentPrice : 월세
roomFacilities : 내부시설
buildingFacilityCodes : 보안 및 기타시설
exposureAddress : 소재지
floorLayerName : 복층 여부
latitude : 위도
longitude : 경도
"""
```

```python
df_sample = df4.loc[:,['exposureAddress','articleFeatureDes','floorInfo','warrantPrice','dealPrice','rentPrice','moveInPossibleYmd','floorLayerName','direction','roomCount','bathroomCount','parkingCount','realtorName','representativeTelNo','realestateTypeName','articleNo','tradeTypeName']]
df_sample.reset_index(inplace = True)
df_sample
```

```python
df_sample.info()
```

```python
for i in range(0,131054):
    df_sample.loc[:,'number_of_room'] = str(df_sample.iloc[i,10])+'개/'+str(df_sample.iloc[i,11])+'개'

df_sample
```

```python
data = df_sample.rename(columns = {'exposureAddress':'address'})
```

```python
data =  data.rename(columns = {'articleFeatureDes':'content'})
```

```python
data = data.rename(columns = {'articleFeatureDes':'content'})
data = data.rename(columns = {'floorInfo':'floor'})
data = data.rename(columns = {'moveInPossibleYmd':'move_in_date'})
data = data.rename(columns = {'floorLayerName':'check_duplex'})
data = data.rename(columns = {'parkingCount':'number_of_parking'})
data = data.rename(columns = {'representativeTelNo':'realter_number'})
data = data.rename(columns = {'realestateTypeName':'buildingfield'})
data = data.rename(columns = {'articleNo':'building_name'})
data = data.rename(columns = {'tradeTypeName':'transaction_type'})
data = data.rename(columns = {'dealPrice':'deal_price'})
data = data.rename(columns = {'realtorName':'realter_name'})
data = data.rename(columns = {'rentPrice':'rent_price'})
data = data.rename(columns = {'warrantPrice':'warant_price'})
```
